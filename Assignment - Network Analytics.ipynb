### Problem Statement: -
There are two datasets consisting of information for the connecting routes and flight halt. Create network analytics models on both the datasets separately and measure degree centrality, degree of closeness centrality, and degree of in-between centrality.
●	Create a network using edge list matrix (directed only).
●	Columns to be used in R:

Flight_halt=c("ID","Name","City","Country","IATA_FAA","ICAO","Latitude","Longitude","Altitude","Time","DST","Tz database time")


```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
#reading the data into python
conn_routs = pd.read_csv(r"D:\360 Data Science 360DigiTMG\Day18-Recomondation engine and network anlytics\Datasets_Network Analytics\connecting_routes.csv")
conn_routs
```

```python
conn_routs.drop(conn_routs.columns[6], axis=1, inplace=True)
```

```python
conn_routs.columns = ["flights", "ID", "main Airport", "main Airport ID", "Destintion", "Destination ID", "haults", "machinary"]
```

```python
conn_routs
```

```python
g = nx.Graph()
```

```python
g = nx.from_pandas_edgelist(conn_routs, source='main Airport', target='Destintion')
```

```python
print(nx.info(g))
```

```python
#degree centrality
b = nx.degree_centrality(g)
b
```

```python
pos = nx.spring_layout(g, k=0.15)
nx.draw_networkx(g, pos, node_size = 25, node_color = 'blue')
```

```python
# closeness centrality
closeness = nx.closeness_centrality(g)
print(closeness)
```

```python
## Betweeness Centrality 
b = nx.betweenness_centrality(g) # Betweeness_Centrality
print(b)
```

```python
## Eigen-Vector Centrality
evg = nx.eigenvector_centrality(g) # Eigen vector centrality
print(evg)
```

```python
# cluster coefficient
cluster_coeff = nx.clustering(g)
print(cluster_coeff)
```

```python
# Average clustering
cc = nx.average_clustering(g) 
print(cc)
```
